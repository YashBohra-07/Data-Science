# Data preprocessing

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_raw)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_raw, test_size=0.3, stratify=y_raw)
```

# XGBoost

```python
from xgboost import XGBClassifier
from sklearn.metrics import plot_confusion_matrix, precision_score,recall_score, f1_score, roc_auc_score, plot_roc_curve
```

```python
xgb = XGBClassifier(max_depth=6)
```

```python
xgb.fit(X_train, y_train)
```

```python
y_pred = xgb.predict(X_test)
y_pred_proba = xgb.predict_proba(X_test)
```

```python
xgb.score(X_test, y_test) # accuracy
```

```python
f1_score(y_test, y_pred)
```

```python
plot_confusion_matrix(xgb, X_test, y_test)
```

```python
plot_roc_curve(xgb, X_test, y_test)
```
